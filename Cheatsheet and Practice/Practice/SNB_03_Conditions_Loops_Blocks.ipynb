Exercise 1:

* Input two variables, marks_student(marks that a student in the class obtained), max_marks(maximum marks scored in the class). 
* Write one block of 'if' statement to compare whether the marks scored by the student is valid (i.e. less than or equal to max marks) or not.

Exercise 7:

Write a 'while' loop to print the sum of 10 positive integers as they pass through the loop one by one.


Eg: 1, 1+2=3, 1+2+3=6

```python
sum = 0
x = 1
while x < 11:
  sum += x
  print(sum)
  x = x + 1
```

Exercise 8:

Write a while loop to print the sum of positive integers. Inside the loop check if the sum of postive integers yet is less than 28. If the sum of positive integers at that point is more than 28, print, the sum is more than 28. The iterations should stop at that point.

```python
sum = 0
x = 1
while x < 11:
  sum += x
  if sum < 28:
    print(sum)
  else:
    print("the sum is more than 28")
    break
  x = x + 1
```

Exercise 9:

In Exercise 8, if sum starts getting bigger than 28, start to print difference between sum and the positive integer.

```python
sum = 0
x = 1
while x < 11:
  sum += x
  if sum < 28:
    print(sum)
  else:
    diff = sum - x
    print(diff)
  x = x + 1
```
