### Lists
Let us take a look at lists. A list is a collection of things, enclosed in [ ] and separated by commas. A list is heterogenous, i.e. the data types of its elements need not be the same

  * A rich set of operations are possible with lists
    - append(), extend(), insert()
    - remove(), sort(), reverse()

```
f = [1, 1, 2, 3, 5, 8]
```

```
f.append(13)
print(f)
```

```
f.extend([21, 34, 55, 89])
print(f)
```

```
a = [1, 2 , 5, 4, 3]
a.sort()
print(a)
```

```
studentNames = ["Jhon", "Lewis", "Peter"]
print(studentNames[::-1])
```

### Tuples
  * Immutable, heterogenous
  * enclosed in parentheses
  * separated by comma

```
a = 17
b = 23
print(a, b)
```

```
c = "Hello"
d = "World!"
e = 42 
c, d = d, c
print(c, d)
e, d = d, e
print(e,d)
```

```
a, b, n = 12, 10, 0
x = (12, 10, 0)
print(x[0])
```
