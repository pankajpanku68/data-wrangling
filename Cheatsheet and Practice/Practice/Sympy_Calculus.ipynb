### Solve Algebraic equations using solve function


**Q1.** Find the value of $x$ for the equation, $x^2$-4=0

   Hint: Use solve() to solve algebraic equations. We suppose all equations are equaled to 0, so solving x**2 = 4 translates into the following code:


```python
import sympy as sym

# Use symbols function to declare the variables
x = sym.symbols('x')

# If we make x**2-4 = 0 then X value will be -2 or 2.  
f1 = x**2-4

# In solve(), first argument is an equation and second argument is symbol. It default returns list
sym.solve(f1,x)
```

### Derivatives

Sympy provides **diff** method for finding the derivative of the function

**Power Rule:**

**Q6.** What is the **third** derivative of a function, f(x) = $x^5$

  Hint: Suppose a function, f(x) = $x^2$, then the derivative of the function w.r.t x is $f'(x)$ = $2x$
        
        

```python
#Define function
f6 = x**5
 
# Diffrentiating exp with respect to the symbol x
# first derivative
der1_x = sym.diff(f6, x) 
print('First derivative w.r.t x:')
print(der1_x)

# YOUR CODE HERE for Second and Third derivative
der3_x = sym.diff(f6, x, 3)
print(der3_x)
```

```python
#Define function
f6 = x**5
 
# Diffrentiating exp with respect to the symbol x
# first derivative
der1_x = sym.diff(f6, x) 
print('First derivative w.r.t x:')
print(der1_x)

# YOUR CODE HERE for Second and Third derivative
der3_x = sym.diff(f6, x, x, x)
print(der3_x)
```

### Limits



To calculate limits in Python, we use the syntax: 

    sympy.limit(function,variable,value)


SymPy can compute symbolic limits with the limit function. The syntax to compute the equation

   \begin{equation}
        limit_{x \to a } = f(x)\\
        is \\
        limit(f(x), x, a)
    \end{equation}


    
                                                          


**Q16.** Compute the limit of the given function,
