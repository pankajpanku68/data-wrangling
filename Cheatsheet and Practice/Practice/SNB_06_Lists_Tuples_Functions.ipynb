#### Exercise 01: Build a function that takes the below string as a parameter and returns the words in the given string in backward order.

Hint: [Split the words and join them in the reverse order using ' '.join(iterable)](https://docs.python.org/3/library/stdtypes.html)

```
string1 = "my name is johnny"
```

```
## YOUR CODE HERE
def rev_sentence(sentence):
    words = sentence.split(' ')
    reverse_sentence = ' '.join(words[::-1])    
    return reverse_sentence

rev_sentence(string1)
```

#### Exercise 06: For the given tuple of numbers, do the following:
* Write a function which takes an integer 'N' as a parameter and returns the sum of multiples of 3 or 5 between 0 and N (limit). Where N is any element from the given tuple.
* Print all the sum of multiples for each element from the tuple.
* Example: If we list all the natural numbers less than 10, that are multiples of 3 or 5, we get 3, 5, 6, and 9 and their sum is 23

```
tuple2 = (10, 20, 30, 40, 50)
```

```
## YOUR CODE HERE
def getMultiples(N):
  sum = 0
  for i in range(1,N):
    if i%3==0 or i%5==0:
      sum += i
  return sum

[getMultiples(each) for each in tuple2]
```

#### Exercise 07: For the given list of sides of a right-angle triangle build a function to extract the length of the hypotenuse and print the hypotenuse values for all parameters.

$ hypotenuse (c) = \sqrt{(a)^{2} + (b)^{2}}$

**Hint:** Use math.sqrt() function from `math` module.

```
import math
parameters = [[4,3],[12,5],[7,24],[8,15],[9,40]]
```

```
## YOUR CODE HERE
def getHypotenuse(a,b):
  return int(math.sqrt(a**2+b**2))

[getHypotenuse(i[0],i[1]) for i in parameters]
```

#### Exercise 08: Validate the below-given passwords based on the provided rules and return the indices of the valid passwords:

* The password must include any special characters ( e.g. @#$%... ) and should have at least one number.

Hint: [String Module](https://docs.python.org/3/library/string.html), [any(iterable)](https://docs.python.org/3/library/functions.html#any)

#### Exercise 09: Build a function to print prime numbers within the given range.

```
## YOUR CODE HERE
limit = 20

[x for x in range(2, limit) if all(x % y != 0 for y in range(2, x))]
```

#### Exercise 10: Rock Paper Scissors: Complete the function below that takes the given input word as a parameter and selects a random element from the words list. It then prints the random word within the function and returns a string, declaring 'win' or 'lose' or 'tie'.

Example: Random input= 'Scissors', User input= Paper, Output = 'lose'

Hint: [random.choice](https://docs.python.org/3/library/random.html)

```
import random
words = ['Rock','Paper','Scissors']
input  = 'Paper'

def game(input):
  ## Complete the function
  print(f"Input was: {input}")

  if input not in words:
    print("Enter any one input among [Rock, Paper, Scissors].")
  
  else: 
    choice = random.choice(words)
    print('The random choice was:',choice)  
    if choice == input:
        print('Its a Tie')
    elif input == 'Rock':
        if choice == 'Paper':
            print('You Lose')
        else:
            print('You Win')

    elif input == 'Paper':
        if choice == 'Rock':
            print('You Win')
        else:
            print('You Lose')

    elif input == 'Scissors':
        if choice == 'Rock':
            print('You Lose')
        else:
            print('You Win')

    else:
        pass

    return None

game(input)
```
